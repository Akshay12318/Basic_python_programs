# Intraday stock volatility analysis

```python
#importing libraries for data analysis

import pandas as pd
import numpy as np

#loading data in pandas dataframe
df=pd.read_excel("Nifty_50.xlsx")
df.columns = df.columns.str.strip()   #for removing extra spaces & gaining clean data from excel

```

```python
#volatility calculation & insertion of new series -Volatility_percentage for analysis

df["Volatility_percentage"]=(((df["HIGH"]-df["LOW"])/df["CLOSE"])*100)
```

```python
df
```

```python
#arrange by descending order of Volatility_percentage

df = df.sort_values(by="Volatility_percentage", ascending=False)
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.lineplot(df["Volatility_percentage"])
```
